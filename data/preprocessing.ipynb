```python
import pandas as pd
```

```python
#https://grouplens.org/datasets/movielens/1m/

data_path = "ratings.parquet"
data = pd.read_parquet(data_path)
```

```python
data = data[data["rating"]>=4.0]
data.drop("rating", axis = 1, inplace = True)
data
```

```python
print(f"Total records : {len(data)}")
print(f"Users : {data['user'].nunique()}")
print(f"Items : {data['item'].nunique()}")
print(f"sparsity ratio : {1-len(data)/(data['user'].nunique()*data['item'].nunique()):.5f}")
```

```python
#filter users watching at least 21 movies
items_per_user = data.groupby("user").size()
data = data[data["user"].isin(items_per_user[items_per_user>=21].index)]
data
```

```python
print(f"Total records : {len(data)}")
print(f"Users : {data['user'].nunique()}")
print(f"Items : {data['item'].nunique()}")
print(f"sparsity ratio : {1-len(data)/(data['user'].nunique()*data['item'].nunique()):.5f}")
```

```python
uid2idx = {uid: i for i, uid in enumerate(data.user.unique())}
mid2idx = {mid: i for i, mid in enumerate(data.item.unique())}

data["user_id"] = data["user"].map(uid2idx)
data["item_id"] = data["item"].map(mid2idx)
```
